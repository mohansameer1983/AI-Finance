#Extract Data

```python
def load_sp_data_from_wikipedia():

  # There are 2 tables on the Wikipedia page
  # we want the first table

  payload=pd.read_html('https://en.wikipedia.org/wiki/List_of_S%26P_500_companies')
  first_table = payload[0]
  second_table = payload[1]

  df = first_table
  symbols = df['Symbol'].values.tolist()
  print(symbols[:15])
  return df, symbols

df, symbols=load_sp_data_from_wikipedia() 
```

```python
df.head()
```

## Get Market Cap

```python
def get_marketcap():
  marketcap_dict={}
  for symbol in df['Symbol']:
      sys=symbol.replace('.', '-')  #BRk.B --> BRK-B
      try:
        info = yf.Ticker(sys).info
        marketcap_dict[symbol] = info['marketCap']
        print(symbol, 'Market Cap: ', info['marketCap'])
      except KeyError:
         print(symbol, 'Error')
  return marketcap_dict

marketcap_dict = get_marketcap()
```

```python
marketcap_dict
```

```python
# Convert dictionary to dataframe
final_info_df = pd.DataFrame(
    {"Symbol": marketcap_dict.keys(), "MarketCap": marketcap_dict.values()}
)
```

# CSV Operations

```python
# Save dataframe to csv
def save_dataframe_to_csv(dataframe, file_path):
    dataframe.to_csv(DATA_PATH + os.sep + file_path, sep='\t')
    return

save_dataframe_to_csv(final_info_df,'/Finance/marketcap.csv')
```

```python
def load_csv_to_dataframe(file_path):
  df = pd.read_csv (DATA_PATH + os.sep + file_path)
  return df

marketdf = load_csv_to_dataframe('/Finance/marketcap.csv')
```
