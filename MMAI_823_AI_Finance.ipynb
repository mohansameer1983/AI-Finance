#Extract Data

```python
def load_sp_data_from_wikipedia():

  # There are 2 tables on the Wikipedia page
  # we want the first table

  payload=pd.read_html('https://en.wikipedia.org/wiki/List_of_S%26P_500_companies')
  first_table = payload[0]
  second_table = payload[1]

  df = first_table
  symbols = df['Symbol'].values.tolist()
  print(symbols[:15])
  return df, symbols

df, symbols=load_sp_data_from_wikipedia() 
```

## Get Market Cap

```python
def get_marketcap():
  marketcap_dict={}
  for symbol in df['Symbol']:
      sys=symbol.replace('.', '-')  #BRk.B --> BRK-B
      info = yf.Ticker(sys).info
      marketcap = info['marketCap']    
      marketcap_dict[symbol] = marketcap
      print(symbol, 'Market Cap: ', marketcap)

  final_info_df = pd.DataFrame.from_dict(marketcap_dict, columns=['Symbol','MarketCap'])
  return final_info_df

market_cap_df = get_marketcap()
```
